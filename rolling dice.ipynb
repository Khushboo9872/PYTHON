```python
#ludo

import random
s=0
t=0
count=0
start=input("want to play y/n:")
while start=='y':
    user=random.randint (1,6)
    print("user:")
    if(user==1):
        print("------\n")
        print("| 0 |\n")
        print("------\n")
    elif(user==2):
        print("------\n")
        print("| 00 |\n")
        print("------\n")
    elif(user==3):
        print("---------\n")
        print("| 0 0 0 |\n")
        print("|-------|\n")
    elif(user==4):
        print("-------\n")
        print("| 0 0 |\n")
        print("| 0 0 |\n")
        print("-------\n")
    elif(user==5):
        print("---------\n")
        print("| 0 0 0 |\n")
        print("|  00   |\n")
        print("---------\n")
    elif(user==6):
        print("-------\n")
        print("| 000 |\n")
        print("| 000 |\n")
        print("-------\n")
    t=t+user
    comp=random.randint (1,6)
    print("comp:")
    s=s+comp
    if(comp==1):
        print("------\n")
        print("| 0 |\n")
        print("------\n")
    elif(comp==2):
        print("------\n")
        print("| 00 |\n")
        print("------\n")
    elif(comp==3):
        print("----------\n")
        print("|  0 0 0  |\n")
        print("|--------|\n")
    elif(comp==4):
        print("-------\n")
        print("| 0 0 |\n")
        print("| 0 0 |\n")
        print("-------\n")
    elif(comp==5):
        print("-------- \n")
        print("| 0 0 0 |\n")
        print("|  00   |\n")
        print("---------\n")
    elif(comp==6):
        print("-------\n")
        print("| 000 |\n")
        print("| 000 |\n")
        print("-------\n")
    count=count+1
    start=input("want to play more?(y/n)..")
    if start=='n':
        print("total user:",t)
        print("total comp:",s)
        print("you played",count,"times")
        


if(t>s):
        print("user is winner")
elif(t<s):
        print("computer is winner")
elif(t==s):
        print("tie")
else:
    print("end")

print("thank you")





```
